```python
import numpy as np

# Sample RNN class to implement a basic RNN
class SimpleRNN:
    def __init__(self, input_size, output_size, hidden_size):
        # Random initialization of weights
        self.Wh = np.random.randn(hidden_size, hidden_size)  # Hidden state weights
        self.Wx = np.random.randn(input_size, hidden_size)   # Input weights
        self.Wy = np.random.randn(hidden_size, output_size)  # Output weights
        self.bh = np.zeros((1, hidden_size))                 # Hidden bias
        self.by = np.zeros((1, output_size))                 # Output bias

    def forward(self, inputs):
        h = np.zeros((inputs.shape[0], self.Wh.shape[0]))  # Initialize hidden state
        outputs = []
        for t in range(inputs.shape[1]):  # For each time step
            h = np.tanh(np.dot(inputs[:, t, :], self.Wx) + np.dot(h, self.Wh) + self.bh)  # Hidden state update
            y = np.dot(h, self.Wy) + self.by  # Output calculation
            outputs.append(y)
        return np.array(outputs)

# Test the RNN
rnn = SimpleRNN(input_size=10, output_size=2, hidden_size=5)
inputs = np.random.randn(3, 5, 10)  # Example batch of 3 sequences, each of length 5, with 10 features
outputs = rnn.forward(inputs)
print("Outputs: ", outputs)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Define the model
model = Sequential([
    Embedding(input_dim=10000, output_dim=32),  # Embedding layer
    SimpleRNN(32),  # Simple RNN layer
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the dataset
max_features = 10000  # Only consider the top 10k words in the dataset
max_len = 500  # Truncate sequences to a maximum length of 500

# Split the data into train and test sets
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Pad sequences to ensure uniform input length
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)

print("Training data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_split=0.2)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc}")
```

```python
import matplotlib.pyplot as plt

# Plot accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```
